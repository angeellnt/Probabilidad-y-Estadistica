```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

media_acc_semana = 2

prob_0_acc = poisson.pmf(0, media_acc_semana)

media_2_semanas = media_acc_semana * 2
prob_3_o_menos = poisson.cdf(3, media_2_semanas)

print(f"Probabilidad de 0 accidentes en 1 semana: {prob_0_acc}")
print(f"Probabilidad de máximo 3 accidentes en 2 semanas: {prob_3_o_menos}")

x = np.arange(0, 10)
y = poisson.pmf(x, media_acc_semana)
plt.bar(x, y, color='purple', alpha=0.6)
plt.title('Distribución de Poisson (media=2)')
plt.show()
```
