```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

monto = [0, 10, 20, 50]
prob = [0.50, 0.30, 0.15, 0.05]

verificacion = sum(prob)
esperanza = sum(x * p for x, p in zip(monto, prob))
varianza = sum(((x - esperanza)**2) * p for x, p in zip(monto, prob))
desviacion = varianza**0.5
acumulada = np.cumsum(prob)

print(f"Suma de probabilidades: {verificacion}")
print(f"Valor esperado E(X): {esperanza}")
print(f"Varianza: {varianza}")
print(f"Desviación estándar: {desviacion}")
print(f"F(x): {acumulada}")

plt.step(monto, acumulada, where='post')
plt.show()

n = 16
p_vacuna = 0.83

prob_ninguno = binom.pmf(16, n, p_vacuna)
prob_todos = binom.pmf(0, n, p_vacuna)

print(f"P(Ninguno enferme): {prob_ninguno}")
print(f"P(Todos enfermen): {prob_todos}")
```
