```python
import numpy as np
from scipy.stats import binom

n = 10
p = 0.10
x = 2

probabilidad_x_2 = binom.pmf(x, n, p)

print("La probabilidad de que la variable aleatoria X sea igual a 2:", probabilidad_x_2)
```
