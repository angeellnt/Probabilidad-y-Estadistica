```python

import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
p = [0.45, 0.30, 0.15, 0.07, 0.03]

suma_p = sum(p)
print(f"Suma de probabilidades: {suma_p}")

esperanza = sum(xi * pi for xi, pi in zip(x, p))
print(f"Valor esperado E(X): {esperanza}")

varianza = sum(((xi - esperanza)**2) * pi for xi, pi in zip(x, p))
desviacion = varianza**0.5
print(f"Varianza: {varianza}")
print(f"Desviación estándar: {desviacion}")

acumulada = []
suma_temp = 0
for pi in p:
    suma_temp += pi
    acumulada.append(suma_temp)

print(f"Distribución acumulada F(x): {acumulada}")

plt.step(x, acumulada, where='post')
plt.title('Función de Distribución Acumulada F(x)')
plt.xlabel('Células anormales')
plt.ylabel('F(x)')
plt.grid(True)
plt.show()
```
