```python
import numpy as np
import matplotlib.pyplot as plt

x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

f = np.cumsum(p)

for xi, fi in zip(x, f):
    print(f"F({xi}) = {fi:.2f}")

plt.step(x, f, where='post')
plt.title("FDA")
plt.xlabel("X")
plt.ylabel("F(x)")
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
