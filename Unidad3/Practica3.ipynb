```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5, 6, 7, 8, 9, 10])
p = np.array([0.05, 0.10, 0.20, 0.30, 0.25, 0.10])

mu = np.sum(x * p)
ex2 = np.sum(x**2 * p)
var = ex2 - mu**2
ds = np.sqrt(var)

print(f"E(X) = {mu:.2f}")
print(f"E(X^2) = {ex2:.2f}")
print(f"Varianza = {var:.2f}")
print(f"Desviación estándar = {ds:.2f}")

plt.stem(x, p, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title("Distribución de Probabilidad")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.grid(True, linestyle='--', alpha=0.6)

plt.axvline(x=mu, color='red', linestyle='--', label=f"Media: {mu:.2f}")
plt.axvspan(mu - ds, mu + ds, color='red', alpha=0.2, label=f"1 Desv: {ds:.2f}")

plt.legend()
plt.ylim(0, max(p) + 0.05)
plt.xticks(x)
plt.show()
```
