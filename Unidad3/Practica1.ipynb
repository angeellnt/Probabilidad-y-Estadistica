```python
import math

x = [5, 6, 7, 8, 9, 10]
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]

ex = 0
for i in range(len(x)):
    ex += x[i] * p[i]

ex2 = 0
for i in range(len(x)):
    ex2 += (x[i] ** 2) * p[i]

var = ex2 - ex**2
ds = math.sqrt(var)

print(f"E(X) = {ex:.2f}")
print(f"E(X^2) = {ex2:.2f}")
print(f"Varianza = {var:.2f}")
print(f"Desviación estándar = {ds:.2f}")
```
