```python
import matplotlib.pyplot as plt

m = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]

suma_p = sum(p)
print(f"Suma de probabilidades: {suma_p}")

esperanza = sum(mi * pi for mi, pi in zip(m, p))
print(f"Ganancia promedio esperada E(X): {esperanza}")

varianza = sum(((mi - esperanza)**2) * pi for mi, pi in zip(m, p))
desviacion = varianza**0.5
print(f"Varianza: {varianza}")
print(f"Desviación estándar: {desviacion}")

acumulada = []
suma_temp = 0
for pi in p:
    suma_temp += pi
    acumulada.append(suma_temp)

print(f"Distribución acumulada F(x): {acumulada}")

plt.step(m, acumulada, where='post')
plt.title('Función de Distribución Acumulada F(x) - Ejercicio 2')
plt.xlabel('Monto del premio (pesos)')
plt.ylabel('F(x)')
plt.grid(True)
plt.show()
```
