```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
p = [0.05, 0.10, 0.25, 0.30, 0.20, 0.10]

suma_p = sum(p)
print(f"1. Verificación de suma: {suma_p}")

esperanza = sum(xi * pi for xi, pi in zip(x, p))
print(f"3. Valor esperado E(X) (promedio de ventas): {esperanza}")

varianza = sum(((xi - esperanza)**2) * pi for xi, pi in zip(x, p))
desviacion = varianza**0.5
print(f"   Varianza: {varianza}")
print(f"   Desviación estándar: {desviacion}")

acumulada = []
suma_temp = 0
for pi in p:
    suma_temp += pi
    acumulada.append(suma_temp)

print(f"4. Función acumulada F(x): {acumulada}")

plt.step(x, acumulada, where='post')
plt.title('Función de Distribución Acumulada - Ventas Semanales')
plt.xlabel('Unidades vendidas')
plt.ylabel('F(x)')
plt.grid(True)
plt.show()
```
