```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos = pd.read_csv("sample_data/autos2.csv", encoding = "ISO-8859-1")
autos.info() #Info tabla
```

```python
print("Valor Maximo precio", autos.price.max())
print("Valor Minimo precio", autos.price.min())
print("Mediana precios:", autos.price.median())
print("Media precios:", autos.price.mean())
print("Desviacion estandar precios:", autos.price.std())
print("")
print("Estadisticos funcion describe:", autos.price.describe())
```

```python
plt.pyplot.hist(autos.price, bins=10)
plt.pyplot.title("Histograma Precios Automoviles")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencia")
```

```python
print("Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil y tipo de vehiculo")
ax = sns.boxplot(x="price", y="body-style", data=autos)
ax.set_ylabel("Tipo de vehiculo")
ax.set_xlabel("Precio")
```

```python
print("Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil y tipo tracción ruedas")
ax = sns.boxplot(x="price", y="drive-wheels", data=autos)
ax.set_ylabel("Traccion de Ruedas")
ax.set_xlabel("Precio")
```

```python
print("Diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil y la ubicacion del motor")
ax = sns.boxplot(x="price", y="engine-location", data=autos)
ax.set_ylabel("Ubicacion del motor")
ax.set_xlabel("Precio")
```

## El precio no se distribuye de manera uniforme y que variables técnicas como la tracción y la ubicación del motor influyen de forma importante en su comportamiento, permitiendo identificar segmentos claros dentro del mercado automotriz representado en el dataset.
