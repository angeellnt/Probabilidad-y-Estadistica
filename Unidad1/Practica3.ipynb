```python
import random
import pandas as pd

edades = []
for i in range(100):
  numero_aleatorio = random.randint(18, 70)
  edades.append(numero_aleatorio)
print(edades)
```

```python
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
tabla_frec = df['Edad'].value_counts().reset_index()
tabla_frec.columns = ['Edad', 'Frecuencia']
tabla_frec = tabla_frec.sort_values(by='Edad')

print("Tabla de distribucion de frecuencias:")
tabla_frec
```

```python
tabla_frec['Frec Relativa'] = tabla_frec['Frecuencia'] / 100
tabla_frec
```

```python
tabla_frec['Frec Porcentual'] = tabla_frec['Frecuencia'] / 100 * 100
tabla_frec
```

```python
tabla_frec['Frec Acumulada'] = tabla_frec['Frecuencia'].cumsum()
tabla_frec
```

```python
import math
print(edades)
df = pd.DataFrame(edades, columns=['Edad'])
df

n =len(edades)
K = int(1 + math.log2(n))
print("Valor de n = ", n, "\nNumero de intervalos =", K)

tabla_frecuencia = pd.cut(df['Edad'], bins=K).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Drec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

print(tabla_frecuencia)
```
