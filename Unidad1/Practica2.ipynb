```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  random_number = round(random.uniform(50, 100), 2)
  fabrica1.append(random_number)
print(fabrica1)

fabrica2 = []
for i in range(100):
  random_number = round(random.uniform(50, 100), 2)
  fabrica2.append(random_number)
print(fabrica2)
```

```python
max1= max(fabrica1)
min1= min(fabrica1)
media1= mean(fabrica1)
mediana1= median(fabrica1)
moda1= mode(fabrica1)
rango1= max1-min1
desviacion1= stdev(fabrica1)

max2= max(fabrica2)
min2= min(fabrica2)
media2= mean(fabrica2)
mediana2= median(fabrica2)
moda2= mode(fabrica2)
rango2= max2-min2
desviacion2= stdev(fabrica2)

print("Estadistica Descriptiva de la fabrica 1 \n", "Media =", media1,
      "\n Mediana =", mediana1, "\n Moda =", moda1, "\n Maximo=", max1,
      "\n Minimo =", min1, "\n Rango =", rango1, "\n Desviacion:", desviacion1)
print("")
print("Estadistica Descriptiva de la fabrica 2 \n", "Media =", media2,
      "\n Mediana =", mediana2, "\n Moda =", moda2, "\n Maximo=", max2,
      "\n Minimo =", min2, "\n Rango =", rango2, "\n Desviacion:", desviacion2)
```

```python
plt.figure(figsize = (8,6))
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={desviacion1:.2f}')
plt.axvline(media1 + desviacion1, color='black', linestyle='dashed', linewidth=2)


plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media2:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana2:.2f}')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda2:.2f}')
plt.axvline(media1 - desviacion2, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={desviacion2:.2f}')
plt.axvline(media1 + desviacion2, color='black', linestyle='dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 2")
plt.legend()
plt.show()
```
