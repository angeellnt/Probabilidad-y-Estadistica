```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  random_number = round(random.uniform(50, 100), 2)
  fabrica1.append(random_number)
print(fabrica1)

fabrica2 = []
for i in range(100):
  random_number = round(random.uniform(50, 100), 2)
  fabrica2.append(random_number)
print(fabrica2)
```

```python
max1= max(fabrica1)
min1= min(fabrica1)
media1= mean(fabrica1)
mediana1= median(fabrica1)
moda1= mode(fabrica1)
rango1= max1-min1
desviacion1= stdev(fabrica1)

max2= max(fabrica2)
min2= min(fabrica2)
media2= mean(fabrica2)
mediana2= median(fabrica2)
moda2= mode(fabrica2)
rango2= max2-min2
desviacion2= stdev(fabrica2)

print("Estadistica Descriptiva de la fabrica 1 \n", "Media =", media1,
      "\n Mediana =", mediana1, "\n Moda =", moda1, "\n Maximo=", max1,
      "\n Minimo =", min1, "\n Rango =", rango1)
print("")
print("Estadistica Descriptiva de la fabrica 2 \n", "Media =", media2,
      "\n Mediana =", mediana2, "\n Moda =", moda2, "\n Maximo=", max2,
      "\n Minimo =", min2, "\n Rango =", rango2)
```
